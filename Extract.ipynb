```python
def clean(df):
    #drop the provinece/state , " lat  and ling columns "
    #select column of index and country/region" only dates in March
    #group by " countryRegion and calculate the sum"
    #reset  the  index of the data fame 
   clean_df = df.drop(columns=['Province/State','Lat','Long']).\
    iloc[:,np.r_[0,40:67]].\
        groupby ('Country/Region').sum().\
            reset_index()
            
   return clean_df
        
    
```

This function is cleaning a dataset by doing a few things:

Removing unnecessary columns

It deletes the columns: 'Province/State', 'Lat', and 'Long' because we donâ€™t need them.
Selecting specific columns

It keeps the first column (which likely contains country names) and some specific columns (from index 40 to 67), which represent dates in March.
Grouping by country

It takes all the rows for the same 'Country/Region' and adds up the numbers (likely COVID-19 cases, population data, or some other metric).
Resetting the index

After grouping, it ensures that the country names are in a neat order instead of being part of the index.

```python
#clean the "confirmed data_df" data set
def convert(clean_df):
    #convert columns of dates into a new column 'Date' and store the data in a new column
    converted =pd.melt(clean_df, id_vars=["Country/Region"],
                       var_name="Date",value_name="Number")
    #convert the data type from string to datetime64 for the data in column 'Date'
    converted['Date'] =pd.to_datetime(converted['Date'])
    #sort the dataset by Country/Region and 'Date'
    converted_df= converted.sort_values(by=["Country/Region","Date"]);
                                        
    return converted_df
```

pd.melt() coreectly reshaped the dataset
pd.to.datetime() converts date values to datetime format.
sort_value(by=["Country/Region","Date"]) sorts the dataFrame properly 
 Briefy explaining 
 Reshape with pd.melt():- This function converts the data from a wide format (where each data is a seperate column) to a long format (with a single "Date" column and one row per date record). We use an identifier column (like "Country/Region") as an id_var.

```python
#apply convert function
confirmed_convert_df =convert(confirmed_clean_df)
recovered_convert_df=convert(recovered_clean_df)
death_convert_df = convert(death_clean_df)

death_convert_df
```

```python
#create a function to calculate the daily increase 
def dailychange(convert_df) :
    #change the index in order to apply dif() function
 convert_df = convert_df.set_index(["Country/Region","Date"])

#Create a new column to store the difference between row 
 convert_df['amount_of_increase'] = convert_df.diff()

#change back the index
 convert_df=convert_df.reset_index()

##Run a for loop to check the boundary rows where the 'Country' changes, and change
 for i in range(1, len(convert_df)):
        if convert_df["Country/Region"][i] != convert_df["Country/Region"][i -1]:
            convert_df.at[i, 'amount_of_increase'] = 0
 convert_df['amount_of_increase'] = convert_df['amount_of_increase'].apply(lambda x: x if x >= 0 else 0) # puttiing the condition where if nbumber is less than 0 keep 0
 convert_df.fillna(0, inplace=True)
 convert_df['amount_of_increase'] = convert_df['amount_of_increase'].astype(int)
 return convert_df
    
```

import pandas as pd

def dailychange(convert_df):
    # Sort the data to make sure it's in order
    convert_df = convert_df.sort_values(by=["Country/Region", "Date"])
    
    # Find the difference in the "Number" column for each country
    convert_df["amount_of_increase"] = convert_df.groupby("Country/Region")["Number"].diff()
    
    # Fill in any missing values with 0
    convert_df["amount_of_increase"] = convert_df["amount_of_increase"].fillna(0)
    
    return convert_df

```python
most_confirmed = pd.read_sql("""
                                SELECT country, sum(increase_of_confirmed) AS total_confirmed
                                FROM confirmed_cleaned_df
                                GROUP BY country
                                ORDER BY SUM(increase_of_confirmed) DESC
                                LIMIT 5
                                """, con=engine)
most_confirmed
```

```python
#converting the result fromsql query into datafram
most_confirmed_df =pd.DataFrame(most_confirmed)
#getting the data into dataframe

#plot bar graph
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax =plt.subplots()
#think of fig as the paper and ax as the space on the paper where you draw your chart
ax.bar(most_confirmed_df['country'], 
       most_confirmed_df['total_confirmed'],
       color='blue',alpha=0.7,
       align="center")
#This is like drawing blue sticks for each country
 #--The height of the stick= number of cases
 #--The name of the country goes under each stick
ax.set_xlabel("Country")
ax.set_ylabel("Number of confirmed cases")
ax.set_title("Top 5 countries with the most confirmed cases in March");
#bottom(x-axis)=Country
#side(y-axis)=Number of cases
#name of the picture=The title
ax.set_xticks(range(len(most_confirmed_df['country'])))
ax.set_xticklabels(most_confirmed_df["country"], rotation=45)
#This makes sure the country names are there and are angled(rotated) so they don't bump into each other


#save fig making the folder to save all the imagines

import os
if not os.path.exists("Image"):
    os.makedirs("Image")
path = os.path.join("Image","most_confirmed.png") #main  this line will send to this imagine to the folder
#path
fig.savefig(path,bbox_inches ='tight')

plt.tight_layout()
plt.show()
```
